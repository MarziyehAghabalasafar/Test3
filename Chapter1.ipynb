<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">1.1 <span style="color:#F50057;">مقدمه</span></h1> <a id='2'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%"> 
به پایتون خوش آمدید،
    یکی از پرکاربردترین زبان های برنامه نویسی کامپیوتری در جهان و بر اساس شاخص محبوبیت زبان های برنامه نویسی 
    (PYPL)
    ، محبوب ترین زبان های جهان است.
 در این فصل، اصطلاحات و مفاهیمی را معرفی می‌کنیم که پایه و اساس یادگیری برنامه‌نویسی پایتون محسوب می‌شوند.
در فصل‌های 2 تا 10 برنامه‌نویسی و در فصل‌های 11 تا 16 موارد مطالعاتی بر مبنای داده، هوش مصنوعی و رایانش ابری ارائه می‌شود.
</div>


<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%"> 
اصطلاحات و مفاهیم برنامه نویسی شیء‌گرا را بررسی خواهیم کرد. خواهید آموخت که چرا پایتون تا این حد محبوب شده است. کتابخانه استاندارد پایتون و کتابخانه‌های مختلف علوم داده را معرفی خواهیم کرد که به شما کمک می‌کنند از نوشتن مجدد کد اجتناب کنید. از این کتابخانه‌ها برای ایجاد اشیاء نرم‌افزاری استفاده خواهید کرد و با کمک این اشیاء کارهای زیادی را با حداقل دستورات انجام خواهید داد.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
در ادامه با سه روش اجرای کدهای پایتون آشنا خواهید شد :
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <span style="color:#8e24aa;">■ </span>در روش اول، از IPython برای اجرای تعاملی دستورالعمل‌های پایتون استفاده کرده و بلافاصله نتایج آن‌ها را مشاهده خواهید کرد.<br>
    <span style="color:#8e24aa;">■ </span>در روش دوم، یک برنامه نسبتا بزرگ خواهید نوشت و این برنامه خلاصه نتایج 6 عدد در پرتاب تاس را به صورت نمودار میله‌ای متحرک نشان می‌دهد. در این برنامه با قانون اعداد بزرگ در عمل آشنا خواهید شد. در فصل 6 این برنامه را با استفاده از کتابخانه  matplotlib ایجاد خواهید کرد.<br>
    <span style="color:#8e24aa;">■ </span>در آخرین روش ژوپیتر نوت‌بوک را معرفی خواهیم کرد. با استفاده از JupyterLab که یک ابزار تحت وب محسوب می‌شود به‌راحتی می‌توانید برنامه خود را نوشته و دستورات پایتون را اجرا کنید. ژوپیتر نوت‌بوک به شما این امکان را می‌دهد که در برنامه خود متن، عکس، صدا، فیلم، انیمیشن، و کد بگنجانید.<br>

</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    در گذشته، بیش‌تر برنامه‌های کامپیوتری روی کامپیوترهای مستقل اجرا می‌شدند
    (یعنی باهم شبکه نبودند). 
    برنامه‌های امروزی را می‌توان باهدف برقراری ارتباط بین میلیاردها کامپیوتر از طریق اینترنت نوشت. مفاهیم رایانش ابری و اینترنت اشیاء را معرفی خواهیم کرد و زمینه لازم برای نوشتن برنامه‌های امروزی در فصول 11 تا 16 را برای شما ایجاد خواهیم کرد.
        در ادامه یاد می‌گیرید که کلان‌داده‌ها چقدر بزرگ هستند و با چه سرعتی رشد می‌کنند. در مرحله بعدی، یک مطالعه موردی کلان‌داده‌ها را در مورد برنامه مسیریاب ویز تلفن همراه ارائه خواهیم داد.این برنامه با استفاده از بسیاری از فن‌آوری‌های مدرن، یک مسیر رانندگی پویا را ارائه می‌دهد که با سریع‌ترین زمان و بهترین حالت به مقصد برسید. همان‌طور که در مورد این فناوری‌ها صحبت می‌کنیم، موارد استفاده آن‌ها در این کتاب را ذکر خواهیم کرد. این فصل با اولین آشنایی ما به بخش علوم‌داده که در حقیقت یک فصل اشتراک کلیدی بین علوم کامپیوتر و هوش مصنوعی محسوب می‌شود، خاتمه می‌یابد.
</div>

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">1.2 <span style="color:#F50057;">مرور سریع اصول فناوری اشیاء</span></h1> <a id='3'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    از آن‌جایی که تقاضا برای نرم‌افزارهای جدید و قدرت‌مند در حال افزایش است، ساختن نرم‌افزار با سرعت، با عملکرد صحیح و البته اقتصادی اهمیت زیادی دارد. اشیاء و یا به عبارت دقیق‌تر، کلاس‌هایی که اشیاء از آن‌ها ایجاد می‌شوند، اساسا اجزای نرم‌افزاری قابل‌استفاده مجدد هستند. اشیاء تاریخ، اشیاء زمان، اشیاء صدا، اشیاء ویدئو، اشیاء اتومبیل، اشیاء افراد، و غیره نمونه‌هایی از اشیاء هستند. تقریبا هر اسمی از‌نظر ویژگی‌ها (به‌عنوان مثال: نام، رنگ و اندازه) و رفتارها (به‌عنوان مثال: محاسبه، جابجایی و برقراری ارتباط) می‌تواند به‌طور معقول به‌عنوان یک شیء نرم‌افزاری نشان داده شود. گروه‌های توسعه‌دهنده نرم‌افزار می‌توانند از رویکرد ماژولار طراحی و پیاده‌سازی شیء‌گرا استفاده کنند. رویکرد اخیر از روش‌های معروف قبلی مانند برنامه‌نویسی ساخت‌یافته بسیار کارآمدتر می‌باشد. درک، برطرف کردن خطا و اصلاح برنامه‌های شیء‌گرا اغلب آسان‌تر است. 
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">اتومبیل به‌عنوان یک شیء</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    برای کمک به شما در درک اشیاء و محتوای آن‌ها، بیایید با یک قیاس ساده شروع کنیم. فرض کنید با اتومبیل رانندگی کرده و با فشار دادن پدال گاز سرعت آن را بیش‌تر کنید. قبل از انجام این کار چه اتفاقی می‌افتد؟ خوب، قبل از راندن یک ماشین، باید یک شخص اتومبیل را طراحی کند. یک اتومبیل به‌طور‌معمول با عنوان نقشه‌های مهندسی شبیه نقشه‌های طراحی خانه شروع می‌شود. پدال گاز جزئیات پیچیده مکانیسم افزایش سرعت اتومبیل را از راننده پنهان می‌کند، به همین نحو پدال ترمز و فرمان اتومبیل جزئیات مکانسیم‌های کاهش سرعت و پیچیدن اتومبیل را از راننده پنهان می‌کند. این رویکرد برای افرادی که دانشی کم و یا کسانی که هیچ‌گونه دانشی در مورد موتور، ترمز و فرمان اتومبیل ندارند، این امکان را فراهم می‌کند تا با اتومبیل به‌راحتی رانندگی کنند.
        همان‌طور که نمی‌توانید وعده‌های غذایی را در طرح اولیه آشپزخانه بپزید، نمی‌توانید با نقشه‌های مهندسی اتومبیل رانندگی کنید. قبل از این‌که بتوانید با ماشین رانندگی کنید، باید از نقشه‌های مهندسی که ماشین را توصیف می‌کنند یک اتومبیل ساخته شود. یک اتومبیل کامل یک پدال گاز واقعی برای افزایش سرعت دارد، اما حتی این هم کافی نیست و اتومبیل به‌خودی‌خود شتاب نمی‌گیرد (امیدوارم!) بنابراین راننده باید پدال را فشار دهد تا ماشین شتاب بگیرد.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">متدها و کلاس‌ها</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    بیایید از مثال اتومبیل برای معرفی برخی مفاهیم کلیدی برنامه‌نویسی شیء‌گرا استفاده کنیم. انجام یک کار در یک برنامه به یک متد نیاز دارد. در این متد کدهایی قرار دارند تا وظیفه آن متد را انجام دهند. همان‌طور که پدال گاز یک ماشین مکانیسم‌های افزایش سرعت اتومبیل را از دید راننده پنهان می‌کند، یک متد این دستورات را از کاربر پنهان می‌کند. در پایتون، یک واحد برنامه به نام کلاس، مجموعه‌ای از متدها دارد که وظایف کلاس را انجام می‌دهند. به‌عنوان‌مثال، کلاسی که نشان‌دهنده یک حساب بانکی است ممکن است حاوی یک متد برای واریز پول به حساب، متد دیگری برای برداشت پول از حساب و متد سومی برای تعیین موجودی حساب باشد. یک کلاس از‌نظر مفهومی مشابه نقشه‌های مهندسی اتومبیل است که طراحی پدال گاز، فرمان و ... را در خود جای‌داده است.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">نمونه‌سازی</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    همان‌طور که قبل از راندن یک اتومبیل باید شخصی، از نقشه‌های مهندسی اتومبیل یک اتومبیل بسازد، قبل از این‌که برنامه‌ای بتواند وظایفی را که متدهای کلاس تعریف می‌کنند اجرا کند، باید یک شیء از کلاس ایجاد شود. فرآیند انجام این کار را نمونه‌سازی می‌نامند. از آن به بعد به یک شیء به‌عنوان نمونه‌ای از کلاس مربوطه محسوب می‌شود.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">استفاده مجدد</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">درست همان‌طور که می‌توان از نقشه‌های مهندسی اتومبیل برای تولید تعداد زیادی از ماشین‌ها استفاده مجدد کرد، شما هم می‌توانید بارها از یک کلاس به‌راحتی برای ساخت بسیاری از اشیاء استفاده کنید. استفاده مجدد از کلاس‌های موجود در زمان ساخت کلاس‌ها و برنامه‌های جدید، موجب صرفه‌جویی در زمان و کار می‌شود. از آن‌جایی که کلاس‌ها و اجزای نرم‌افزاری موجود اغلب تحت آزمون‌های زیاد، عیب‌یابی و تنظیم کارایی قرار گرفته‌اند به شما کمک می‌کنند تا سیستم‌های قابل اطمینان و کارآمدتر ایجاد کنید. درست همان‌طور که مفهوم قطعات قابل تعویض برای انقلاب صنعتی بسیار حیاتی بودند، مفهوم کلاس‌های با قابلیت استفاده مجدد برای انقلاب نرم‌افزاری بسیار مهم تلقی می‌شوند و در حقیقت ماحصل فناوری شی‌ء‌گرا هستند.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">پیام‌ها و فراخوانی متدها</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">هنگامی که حین رانندگی با اتومبیل پدال گاز را فشار می‌دهید، در حقیقت پیامی را برای انجام یک وظیفه و یا افزایش سرعت به اتومبیل ارسال می‌کنید. به همین ترتیب، به یک شیء پیام ارسال می‌کنید. هر پیام به‌صورت فراخوانی یک متد اجرا می‌شود و به متد آن شیء می‌گوید تا وظیفه خود را انجام دهد. به‌عنوان‌مثال، ممکن است یک برنامه برای افزایش موجودی شیء
    bank-account
    ، 
    متد
    deposit
    آن را فراخوانی کند.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">صفت‌ها و متغیرهای نمونه‌سازی شده</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">یک اتومبیل علاوه‌بر داشتن قابلیت‌هایی برای انجام وظایف خود، دارای ویژگی‌هایی نظر رنگ، تعداد در، مقدار سوخت موجود در باک، سرعت فعلی آن و ثبت کل مسافت طی شده(به‌عنوان‌مثال، خواندن کیلومتر شمار) است. ویژگی‌های خودرو مانند قابلیت‌های آن به‌عنوان بخشی از طراحی خودرو در نقشه‌های مهندسی نشان داده می‌شوند (به‌عنوان‌مثال شامل کیلومتر شمار و شاخص سوخت است). علاوه‌بر این، هنگامی که با اتومبیل واقعی رانندگی می‌کنید، این ویژگی‌ها همراه با ماشین حمل می‌شوند. هر اتومبیل ویژگی‌های خاص خود را حفظ می‌کند. به‌عنوان‌مثال، هر اتومبیل می‌داند که در باک خود چه مقدار سوخت دارد، اما در مورد مقدار سوخت موجود در باک سایر اتومبیل‌ها چیزی نمی‌داند.
        به همین نحو یک شیء دارای ویژگی‌هایی است که همراه با آن در یک برنامه استفاده می‌شود. این ویژگی‌ها به‌عنوان بخشی از کلاس شیء مشخص می‌شوند. برای مثال، یک شیء حساب بانکی، دارای ویژگی موجودی است که مقدار پول در آن حساب را نشان می‌دهد. هر شیء حساب بانکی موجود خود را می‌داند، اما دو مورد موجودی دیگر حساب‌ها چیزی نمی‌داند. این ویژگی‌ها توسط متغیرهای نمونه‌سازی شده کلاس مشخص می‌شوند. ویژگی‌ها و متدهای یک کلاس (واشیای آن) ارتباط تنگاتنگی با یکدیگر دارند و به همین دلیل کلاس‌ها ویژگی‌ها و متدهای خود را بسته‌بندی می‌کنند.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">وراثت</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">با استفاده از وراثت به‌راحتی می‌توان یک کلاس جدید(به نام کلاس فرزند) را ایجاد کرد. این فرآیند با ویژگی‌های یک کلاس موجود (به نام کلاس والد) شروع شده، احتمالا آن‌ها را شخصی‌سازی کرده و ویژگی‌های منحصربه‌فرد خود را با آن اضافه می‌کند. در مثال اتومبیل، مطمئنا یک شیء از کلاس
    convertible
    یک شیء از کلاس عمومی اتومبیل است، اما به‌طور خاص‌تر، سقف این خودرو قابلیت باز و بسته شدن دارد.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">تحلیل و طراحی شیء‌گرا</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">به‌زودی برنامه‌هایی به زبان پایتون خواهید داشت اما چگونه کد برنامه‌های خود را آماده می‌کنید؟
    شاید مانند بسیاری از برنامه‌نویسان به‌سادگی کامپیوتر خود را روشن کرده و شروع به تایپ کردن کنید. ممکن است این روش برای برنامه‌های کوچک (مانند برنامه‌هایی که در فصل‌های اول کتاب ارائه‌شده‌اند) کارساز باشد، اما اگر از شما خواسته شد یک سیستم نرم‌افزاری ایجاد کنید تا هزاران دستگاه خودپرداز را برای یک بانک بزرگ کنترل کند؟ یا فرض کنید که از شما خواسته شد روی گروهی متشکل از 1000 توسعه‌دهنده نرم‌افزار کار کنید که تا نسل بعدی سیستم کنترل ترافیک هوایی ایالت متحده را بسازند؟ برای پروژه‌های بزرگ و پیچیده، نباید به‌سادگی بنشینید و شروع به نوشتن برنامه کنید.
        برای ایجاد بهترین راه‌حل‌ها، باید یک فرآیند تجزیه‌و‌تحلیل دقیق برای تعیین نیازمندی‌های پروژه خود دنبال کنید(به‌عنوان‌مثال، تعیین کنید که قرار است سیستم چه کاری انجام دهد)، سپس طرحی را تهیه کنید که این نیازمندی‌ها را برآورده می‌کند(به‌عنوان‌مثال، مشخص کنید که سیستم چگونه باید این‌کار را انجام دهد). در حالت ایده‌آل، می‌بایست قبل از نوشتن هر کد، این مراحل را طی کرده و با دقت طراحی را بررسی کنید(و طرح خود را با متخصصان دیگر نرم‌افزار هم مورد بررسی قرار دهید). اگر این فرآیند شامل تجزیه‌و‌تحلیل و طراحی سیستم از دیدگاه شیء‌گرا باشد، فرآیند تحلیل و طراحی شیء‌گرا
    OOAD
    نامیده می‌شود.
    زبان‌هایی مانند پایتون شیء‌گرا هستند. برنامه‌نویسی در چنین زبانی که برنامه‌نویسی شیء‌گرا
    OOP 
    نامیده می‌شود، به شما امکان می‌دهد یک طرح شیء‌گرا را به‌عنوان یک سیستم کاری پیاده‌سازی کنید.
</div>

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">1.3 <span style="color:#F50057;">پایتون</span></h1> <a id='4'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">پایتون یک زبان برنامه‌نویسی است که در سال 1991 به‌صورت عمومی منتشر شد. این زبان توسط خودو فان روسوم از موسسه ملی تحقیقات ریاضیات و علوم‌کامپیوتر آمستردام توسعه یافت. زبان پایتون به یکی از رایج‌ترین زبان‌های برنامه‌نویسی دنیا تبدیل شده است. این زبان در حال حاضر به‌طور خاصی برای مصارف آموزشی و علمی محبوب است و به‌تازگی از زبان برنامه‌نویسی

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">1.4 <span style="color:#F50057;">کتابخانه‌ها</span></h1> <a id='5'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در طول این کتاب، به استفاده از کتابخانه‌های موجود در پایتون تمرکز می‌کنیم تا از کدنویسی دوباره اجتناب کنید و با این کار تلاش شما در توسعه برنامه بهینه می‌شود. غالبا به‌جای نوشتن کد از ابتدا‌به‌ساکن که فرآیندی زمان‌بر و پرهزینه است، بهتر است برای ایجاد یک شیء، از کلاس‌های از قبل تعریف‌شده در کتابخانه این زبان استفاده کنید. این کار فقط با یک دستور در پایتون انجام می‌شود؛ بنابراین، کتابخانه‌ها به شما کمک می‌کنند تا با حداقل کد، کارهای قابل‌توجهی انجام دهید. در این کتاب از طیف گسترده‌ای از کتابخانه‌های استاندارد پایتون، کتابخانه‌های علم‌داده و کتابخانه‌های ثالث استفاده خواهید کرد.

<h2 style="direction:rtl;font-size: 25px;font-family: 'XB Zar';color:#8e24aa;">1.4.1 <span style="color:#8e24aa;">کتابخانه‌های استاندارد پایتون</span></h2> <a id='6'></a>

<h2 style="direction:rtl;font-size: 25px;font-family: 'XB Zar';color:#8e24aa;">1.4.2 <span style="color:#8e24aa;">کتابخانه‌های علم‌داده</span></h2> <a id='7'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">پایتون دارای یک انجمن بزرگ و رو به رشد از توسعه‌دهندگان متن‌باز در بسیاری از زمینه‌ها است. کتابخانه‌های متن‌باز که توسط انجمن متن‌باز پایتون توسعه‌یافته‌اند، یکی از بزرگ‌ترین دلایل محبوبیت این زبان به شمار می‌روند. یکی از اهداف ما ایجاد نمونه‌ها و مطالعات موردی در زمینه پیاده‌سازی است. این کار برای شما یک معرفی جذاب، چالش‌برانگیز و سرگرم‌کننده‌ای در زمینه برنامه‌نویسی پایتون به‌ارمغان می‌آورد و در عین‌حال شما را با علم‌داده، کتابخانه‌های مهم علم‌داده و موارد دیگر درگیر می‌کند. از کارهای اساسی که می‌توانید با چند خط کد انجام دهید شگفت‌زده خواهید شد. در جدول زیر کتابخانه‌های رایج علم‌داده لیست شده‌اند. بسیاری از این کتابخانه‌ها را هنگام کار با مثال‌های علم‌داده مورداستفاده قرار خواهید داد. برای گرافیکی کردن داده‌ها از

    <a href="http://pyviz.org/">http://pyviz.org</a>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">پردازش داده‌های جهان به برق زیادی نیاز دارد</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">داده‌های دستگاه‌های متصل به اینترنت در جهان به طور فزاینده‌ای در حال افزایش است و پردازش آنها به مقدار زیادی انرژی نیاز دارد. طبق مقاله به روز مصرف انرژی برای پردازش داده‌ها در سال ۲۰۱۵ با ۲۰ درصد در سال در حال رشد بود و تقریباً ۳ تا ۵ مصرف درصد انرژی جهان را به خود اختصاص داد. این مقاله می‌گوید که کل مصرف برق پردازش داده تا سال ۲۰۲۵ می‌تواند به ۲۰ درصد برسد. 
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%"> ارز دیجیتالی بیت کوین که با فناوری بلاک چین کار می‌کند، یکی از مصرف کننده‌های زیاد انرژی محسوب می‌شود.مقدار انرژی لازم برای پردازش یک تراکنش بیت کوین تقریباً با میزان مصرف برق یک هفته یک خانواده آمریکایی برابری می‌کند. ماینرها از این انرژی برای اثبات درستی یک تراکنش استفاده می‌کنند. 
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">طبق برخی تخمین‌ها میزان مصرف انرژی تراکنش‌های بیت کوین از انرژی مصرفی سالیانه خیلی کشورها بیشتر است. بیت کوین و اتریوم (یکی دیگر از پلتفرم‌های رایج مبتنی بر بلاک چین و ارز دیجیتال) در سال بیشتر از اسرائیل و تقریباً به اندازه یونان انرژی مصرف می‌کنند. 
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">مورگان استنلی در سال ۲۰۱۸ پیش بینی کرد که مصرف برق مورد نیاز برای ایجاد ارزهای دیجیتالی در سال جاری می‌تواند از مقدار انرژی مورد نیاز جهانی خودروهای الکتریکی که شرکت‌ها برای سال ۲۰۲۵ پیش‌بینی کرده‌اند، بیشتر شود.با توجه به علاقه زیاد به برنامه‌های مبتنی بر بلاک چین که از ارز دیجیتالی هم بیشتر شده این وضعیت پایدار نمی‌ماند. انجمن بلاک چین به دنبال راه‌حلی برای این معضل هستند.

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">فرصت‌های کلان‌داده</h2> <a id=''></a>

<h2 style="direction:rtl;font-size: 25px;font-family: 'XB Zar';color:#8e24aa;">1.7.1 <span style="color:#8e24aa;">تجزیه و تحلیل کلان‌داده</span></h2> <a id='16'></a>

    <span style="color:#D50000;">3. </span><b>تنوع:</b> قبلاً داده‌ها به صورت حرف و عددی بودند (یعنی متشکل از حروف الفبا، ارقام، علائم نگارشی و برخی کاراکترهای خاص)، امروزه داده‌ها شامل تصاویر، فایل‌های ویدیویی و فایل‌های صوتی هستند. علاوه بر این داده‌ها، داده‌های بسیار زیادی از جانب سنسور اشیای مختلف در منازل، کسب‌و‌کارها، ماشین‌ها و شهرها ارسال می‌گردد.<br>
    <span style="color:#D50000;">4. </span><b>درستی:</b> اعتبار داده‌ها - آیا داده‌ها کامل و دقیق هستند؟ آیا می‌توانیم هنگام تصمیم گیری حیاتی به این داده‌ها اعتماد کنیم؟ آیا این داده‌ها واقعی هستند؟<br>
